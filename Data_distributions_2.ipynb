```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
univ_age=4.35e17/((10**6)*365*24*60*60)

BHBH_01_analysed=pd.read_csv("BHBH_n_01.csv")
BHBH_01_analysed=BHBH_01_analysed.loc[BHBH_01_analysed["time_delays_RK"]<univ_age]

BHBH_001_analysed=pd.read_csv("BHBH_n_001.csv")
BHBH_001_analysed=BHBH_001_analysed.loc[BHBH_001_analysed["time_delays_RK"]<univ_age]

BHBH_0001_analysed=pd.read_csv("BHBH_n_0001.csv")
BHBH_0001_analysed=BHBH_0001_analysed.loc[BHBH_0001_analysed["time_delays_RK"]<univ_age]

```

```python
from scipy.optimize import curve_fit

binedges = np.logspace(np.log10(min(np.array(BHBH_01_analysed["time_delays_RK"])/10**3)), np.log10(max(np.array(BHBH_01_analysed["time_delays_RK"])/10**3)), 30)

plt.hist(np.array(BHBH_01_analysed["time_delays_RK"])/10**3, bins=binedges, alpha=0.5,density=True) #in gigayears

# Calcolare la posizione del centro di ciascun bin
hist, edges = np.histogram(np.array(BHBH_01_analysed["time_delays_RK"])/10**3, bins=binedges,density=True)
bin_centers = (edges[:-1] + edges[1:]) / 2


def plaw(x,K,alpha):
    return K/(x**alpha)


popt,pcov=curve_fit(plaw,bin_centers,hist)
xtest=np.linspace(0.0065,13.9,20000)


plt.plot(bin_centers, hist, 'bo', markersize=5)
plt.plot(xtest,plaw(xtest,popt[0],popt[1]),label='power law fit')
plt.xlabel('T_delay in Gyears')
plt.ylabel('Frequenza')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()


plt.hist(np.array(BHBH_01_analysed["time_delays_RK"])/10**3, bins=20, alpha=0.5) #in gigayears

# Calcolare la posizione del centro di ciascun bin
hist, edges = np.histogram(np.array(BHBH_01_analysed["time_delays_RK"])/10**3, bins=20)
bin_centers = (edges[:-1] + edges[1:]) / 2

popt,pcov=curve_fit(plaw,bin_centers,hist)
xtest=np.linspace(0.35,13.9,20000)
plt.plot(bin_centers, hist, 'bo', markersize=5)
plt.plot(xtest,plaw(xtest,popt[0],popt[1]),label='power law fit')
plt.xlabel('T_delay in Gyears')
plt.ylabel('Frequenza')
plt.grid()
plt.legend()
plt.show()
popt[1]
```
