```python
# Carica il datase
def dati(data):
    dataset = np.array(data)

    x = dataset[:, 0:4]# Parametri del sistema
    x[:, 0:2] = x[:, 0:2].astype(float) * (2 * 10 ** 30)
    x[:, 2] = x[:, 3].astype(float) * 6.96e8

    y = dataset[:, 4]*(365 * 24 * 60 * 60 * 10**6)  # Tempo di merge
    N = len(x)
    
    perc_train=0.8
    # dim. of a sample
    L = len(x[0])
    print(L)

    for i in range(10):
        print(x[i],y[i])

    N_train = int(perc_train * N)
   
    print(f'data: {N}\ntrain: {N_train}')
    # Suddivide il dataset in set di addestramento e test

    (x_train, y_train) = (x[0:N_train],y[0:N_train])
    (x_test, y_test) = (x[N_train:],y[N_train:])
    print("Train:",len(x_train),"\t Validation:",len(x_test))
    print(len(x_train))
    return x_train, y_train,x_test, y_test
```
